```python
#---- A program to recognize handwritten numbers or Digits Recognition using keras cnn---
# Here the model was built using the keras built-in mnist dataset
# And i imported handwritten images and passed it to the model to test how well it works
#-------------------------------------
# 1- importing requested library
import tensorflow as tf
from tensorflow import keras
# print(keras.__version__) #checking keras installed properly
```

```python
# 2- Loading data
from keras.datasets import mnist
(x_train,y_train),(x_test,y_test)=mnist.load_data() # Loading mnist datasets in a specific parameters
print('Shape of x_train is : ',x_train.shape)
```

```python
# 3- processing ddta
x_train = x_train.reshape(x_train.shape[0],28,28,1) #reshaping images to 28*28pixels and grayscale on channel 1
x_test = x_test.reshape(x_test.shape[0],28,28,1) #reshaping images to 28*28pixels and grayscale on channel 1
print('Shape of x_train is : ',x_train.shape) # checking rehaped data

# Normalizing the input feature (shape)
x_train=x_train/255.0 # image normalization
x_test=x_test/255.0

# #Categoricaling the output lable
# from keras import utils
# num_classes=10 # the number of out put labels is from 0 to 9 = 10
# y_train=y_train.astype('float32')
# y_train=keras.utils.to_categorical(y_train,num_classes)
# y_test=keras.utils.to_categorical(y_test,num_classes)

from tensorflow import keras
y_train=y_train.astype('float32')
y_test=y_test.astype('float32')
y_train=keras.utils.to_categorical(y_train,num_classes=10,dtype='float32')
y_test=keras.utils.to_categorical(y_test,num_classes=10,dtype='float32')

```

```python
#4- Building the model
# Sequential is a function that stores cnn layers as a stack
from keras.models import Sequential
model=Sequential() # Preparing our model for the input layers
#importing the requested layers
from keras.layers import Conv2D,MaxPool2D,Dropout,Flatten,Dense
# Adding layers to our model
#a- Conv2D input Layer output_shape=input_shape-kernel_size+1
input_shape=(28,28,1)
model.add(Conv2D(32,(8,8),activation='relu',input_shape=input_shape,name='Conv2D_input_layer'))
model.add(Conv2D(64,(5,5),activation='relu',name='conv2d_2'))
#b- MaxPool2D layer divide by 2 output_shape=input/2
model.add(MaxPool2D(2,2))
#c- dropout layer to drop out 0.2 of the numbers of neons to protect from overfitting
model.add(Dropout(0.2))
#d- Fully connected Flatten layer output_shape=input*dims
model.add(Flatten())
#e- Adding Dense and output layer
model.add(Dense(32,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(10,activation='softmax'))# output layer

model.summary()
```

```python
# 5- Compiling the model
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics='accuracy')
print('compiled')
```

```python
# 6- Fitting the model
model.fit(x_train,y_train,batch_size=128,epochs=5,validation_data=(x_test,y_test))
```

```python
# 7- model evaluation And Saving
score = model.evaluate(x_test,y_test,verbose=2)
print('Accuracy is : %',score[1],'\n And Loss is : ',score[0])
model.save('K_mnist_model')
```

```python
# 8- Predicting .......................
# We firstly need to load the image to be predicted
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
image = load_img('../input/handwritten-image/mnist_pic_7.PNG',target_size=(28,28),color_mode = "grayscale")
image=img_to_array(image)
image=image.reshape(1,28,28,1)
image=image.astype('float32')
image=image/255.0

#predicting
from keras.models import load_model
model=load_model('./K_mnist_model')
res=model.predict(image)
print('The array of propabilities for the input image is likely to be :\n',res)# the result now is an array of propabilities

# To pick up the index of the highest probability that express the handwritten digit we use numpy
import numpy as np
predicted_num=np.where(res==np.amax(res))
print("Hey ....... \nThe predicted number of your input image is : ",predicted_num[1])
```
